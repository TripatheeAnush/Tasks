## Linear Algebra Task

1. Matrix Operations

Perform the following operations on the given matrices:

A = [[2, -1],
     [0,  3]]

B = [[1,  2],
     [-3, 4]]

a. A + B  
b. AB  
c. 3A - 2B


```python
import numpy as np
A=np.array([[2,-1],
            [0,3]])
B=np.array([[1,2],
            [-3,4]])
sum=A+B   
product=A@B
result=3*A-2*B
print("A+B=",sum)
print("AB=",product)
print("3A-2B=",result)
```

2. Determinants and Inverses  
Find the determinant of A.
Determine whether A is invertible. If yes, calculate A^−1.

```python
import numpy.linalg as LA
A=np.array([[2,-1],
            [0,3]])
det_A=LA.det(A)
print("Determinant of A=",det_A)
if det_A != 0:
    inv_A=LA.inv(A)
    print("Inverse of A=",inv_A)
else:
    print("Matrix A is not invertible")
```

3. Eigenvalues and Eigenvectors  
Compute the eigenvalues and eigenvectors of the matrix:
C=[[4    1], [​−2       1]​]

```python
C=np.array([[4,1],
            [-2,1]])
eigenvalues,eigenvectors=LA.eig(C)
print("Eigenvalues of C=",eigenvalues)
print("Eigenvectors of C=",eigenvectors)
```

You are given a dataset of cars with the following features:Number of cylinders (x1​)Engine size in liters (x2​)Horsepower (x3)
car 1:
x1: 4
x2:  1.5
x3: 100
and price: 20

and price: 30

Then, find the price of a car 3 with features
x1: 8
x2:  3.0
x3: 200
price:?

Hints:
1. Represent the given dataset as a matrix equation X.T*β=y, where X is the feature matrix, β is the coefficient vector, and y is the price vector  
2. Compute β (the coefficients) using the normal equation:
β=(X.T * X)^−1  X.T y

where,
X.T --> transpose matrix

```python
X=np.array([
    [4,1.5,100],
    [6,2.0,150]
])
Y=np.array([20,30])
XT=X.T
XTX=(X.T)@X
det_XTX=LA.det(XTX)
if det_XTX != 0:
    beta=LA.inv(XTX) @ XT @ Y       #Normal Equation
else:
    beta=LA.pinv(X) @ Y             #Pseudo Inverse Calculation as XTX is singular matrix and has determinant 0
print("Coefficient Vector=",beta)
car3=np.array([8,3.0,200])
car3_T=car3.T
price=car3.T @ beta     # XT * beta = Y
print("Predicted price of Car 3=",price)
```
