```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("Health_heart_experimental.csv")
df.head(5)
```

Line chart


```python
df = pd.read_csv("Health_heart_experimental.csv")
avg_data = df.groupby("age")["HR"].mean()

plt.plot(avg_data.index, avg_data.values, marker='o')
plt.xlabel("Age")
plt.ylabel("Average Maximum Heart Rate")
plt.title("Age vs Average Maximum Heart Rate")
plt.show()
```

The graph illustrates how the average maximum heart rate changes with age. Overall, the values do not follow a strict pattern but fluctuate slightly across different ages. However, younger age groups tend to show a somewhat higher average maximum heart rate compared to older age groups. For instance, the average maximum heart rate around 25 years appears higher than that around 55 years. This suggests that while the changes are not very strong, there is a mild tendency for the average maximum heart rate to decrease as age increases.

Bar Plot

```python
avg_sysbp = df.groupby("sex")["SysBP"].mean()

plt.bar(["Male (0)", "Female (1)"], avg_sysbp)
plt.xlabel("Sex")
plt.ylabel("Average Systolic Blood Pressure")
plt.title("Sex vs Average Systolic Blood Pressure")
plt.show()
```

The bar plot compares the average systolic blood pressure of males (0) and females (1).
It shows that both groups have very similar average blood pressure values, indicating no major difference.
Females have a slightly higher average systolic blood pressure than males, but the difference is minimal.

Pie Chart

```python
df['sex'].value_counts().plot(kind='pie',labels=['Male(0)','Female(1)'],autopct='%0.2f')
```

The pie chart illustrates the distribution of individuals by sex in the dataset. Males, denoted by 0, make up 50.09% of the sample, while females, denoted by 1, account for 49.91%. This indicates that the dataset is nearly evenly balanced between male and female individuals, with only a very small difference in representation.

Box Plot

```python
sns.boxplot(x="indication", y="BMI", data=df)

plt.xlabel("Heart Condition Status (0 = No, 1 = Yes)")
plt.ylabel("Body Mass Index (BMI)")
plt.title("BMI Distribution by Heart Condition Status")
plt.show()
```

The box plot compares BMI values between individuals with and without a heart-related condition. Those without a heart condition generally have lower BMI levels, mostly within the normal to overweight range. In contrast, individuals with a heart condition tend to have higher BMI values. The distribution for the heart-condition group is shifted upward overall, indicating that higher body weight is more common among these individuals. The plot also shows greater variability in BMI among individuals with a heart condition. Also, several outliers appear in the heart-condition group at lower BMI values, highlighting that heart conditions can still occur even among individuals with relatively low BMI. Overall, the box plot suggests a clear association between higher BMI and the presence of a heart-related condition.


Histogram

```python
sns.displot(df['age'], kind="hist", kde=True)
```

The histogram displays the distribution of ages in the dataset, with age on the x-axis and the count on the y-axis. The bars and the smooth density curve indicate that ages are fairly evenly spread between approximately 20 and 70 years. There are no sharp peaks or gaps, suggesting that no single age group dominates the dataset. When comparing the 2030 age group with the 6070 age group, both ranges show similar frequencies. Overall, the dataset includes a well-balanced mix of age groups, supporting reliable age-based comparisons.

Scatter plot

```python
sns.scatterplot(x="SysBP", y="DiaBP", data=df,color='green')

plt.xlabel("Systolic Blood Pressure (mmHg)")
plt.ylabel("Diastolic Blood Pressure (mmHg)")
plt.title("Systolic vs Diastolic Blood Pressure")
plt.show()
```

The scatter plot compares systolic blood pressure (SysBP) and diastolic blood pressure (DiaBP) to examine their relationship. Each point represents an individuals blood pressure reading, with systolic pressure on the x-axis and diastolic pressure on the y-axis. The points form a broadly upward pattern, indicating a positive relationship where higher systolic values generally correspond to higher diastolic values. However, the relationship is not perfectly linear, as there is noticeable spread across the plot. The grid-like appearance occurs because blood pressure measurements are recorded as whole numbers, causing many points to overlap at the same values. Overall, the plot suggests that systolic and diastolic pressures tend to rise together, while still showing individual variability.
